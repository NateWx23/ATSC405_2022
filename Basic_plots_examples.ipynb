In this notebook, there are short examples of code for plotting line, scatter, and bar plots.

First, we will start by making some "toy" data into two lists. We will use these lists for all of the examples. 

```python
sample_1 = [22,43,12,54,33]
sample_2 = [24,43,14,52,32]
```

We will make use of the matplotlib.pyplot module 

```python
import matplotlib.pyplot as plt
```

Now we will plot a simple line plot with both sets of data on the same plot.

```python
#Create the figure object 
#This is what we will add our axis (actual plot with the data) on
fig = plt.figure(figsize=(8,5))
#Create the axes object that is associated with the figure object we just created 
ax = fig.add_axes([0,0,1,1])
#Plot the data on the axes object 
ax.plot(sample_1,color='blue',label='sample 1')
ax.plot(sample_2,color='red',label='sample 2')

#sets the x and y axes limits if you don't want the defaults 
ax.set_xlim(-1,5)
ax.set_ylim(8,60)
#add labels to the plot
ax.set_title('Title of Plot',size=20)
ax.set_xlabel('X label',size=20)
ax.set_ylabel('Y label',size=20)
#turn on the grid 
ax.grid()
#call the legend method on the axes object 
ax.legend(loc='upper left')
```

Next, we will make a scatter plot to directly compare the two datasets. 

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_axes([0,0,1,1])
ax.scatter(sample_1,sample_2,color='red')


ax.set_xlim(10,55)
ax.set_ylim(10,55)


ax.set_title('Title of Plot',size=20)
ax.set_xlabel('Sample 1',size=20)
ax.set_ylabel('Sample 2',size=20)
ax.grid()
```

Finally, let's make a bar plot for one of the datasets.

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_axes([0,0,1,1])

ax.bar(range(5),sample_1,zorder=2)

ax.set_title('Title of Plot',size=20)
ax.set_ylabel('Y label',size=20)
#labels keyword did not work for this method - apparently only works in certain versions of matplotlib
#for this method - had to use set_xticklabels instead of passing the label keyword into xticks 
ax.set_xticks(range(5))
ax.set_xticklabels(['July 01','July 02','July 03','July 04','July 05']) 

ax.grid()
```
